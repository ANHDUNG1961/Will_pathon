Below we will build a rock paper scissors game

```python
def rock_paper_scissors():
  computer_options = ['Rock', 'Paper', 'Scisors']
  #pc_choice = computer_options[random.randrange(0,3)]
  pc_choice = time.localtime()[5]
  if pc_choice%3 == 0:
    pc_choice = computer_options[2]
  elif pc_choice%2 == 0:
    pc_choice = computer_options[1]
  else:
    pc_choice = computer_options[0]
  your_choice = input()
  while your_choice not in computer_options:
    print("I don't understand that choice please try again")
    your_choice = input()

  print(pc_choice)
  if (your_choice, pc_choice) in [('Rock', 'Scisors'),('Scisors', 'Paper'),('Paper','Rock')]:
    print(your_choice+' beats '+pc_choice+' you win! ')
  elif (pc_choice, your_choice) in [('Rock', 'Scisors'),('Scisors', 'Paper'),('Paper','Rock')]:
    print(your_choice+' loses to '+pc_choice+' you lose! ')
  else:
    print("it's a draw!")
```

```python
rock_paper_scissors()
```

```python
def find_blank_space(game_board):
  row = random.randrange(0,3)
  column = random.randrange(0,3)
  if game_board[row][column] == None:
    game_board[row][column] = 'o'
    return None
  elif None not in game_board:
    return None
  else:
    find_blank_space(game_board)
```

```python
import numpy as np
def tic_tac_toe():
  game_board = np.array([[None,None,None],[None,None,None],[None,None,None]])
  while None in game_board:
    your_move = input()
    game_board[int(your_move[0])][int(your_move[2])] = 'x'
    if game_board[1][1] == None:
      game_board[1][1] = 'o'
    elif game_board[1][1] != None:
      loop_variable = True
      for i in range(0,3):
        if loop_variable == False:
          break
        for j in range(0,3):
          if loop_variable == False:
            break
          if game_board[i][j] == 'x' and game_board[i][j-1] == None:
            game_board[i][j-1] = 'o'
            loop_variable = False
          elif game_board[i][j] == 'x' and game_board[i-1][j] == None:
            game_board[i-1][j] = 'o'
            loop_variable = False
    else:
      find_blank_space(game_board)
    print(game_board)
```

```python
tic_tac_toe()
```
