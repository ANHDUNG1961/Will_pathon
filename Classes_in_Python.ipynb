Now an example of a class of object with methods on it that could be interesting from a data analysis perspective: Mortgages.

```python
class Mortgage:
  def __init__(self, years, interest_rate, principal, compounding, house, payment_amount, payment_frequency, up_front_fees):
    self.years = years
    self.interest_rate = interest_rate
    self.principal = principal
    self.compounding = compounding
    self.house = house
    self.payment_amount = payment_amount
    self.payment_frequency = payment_frequency
    self.up_front_fees = up_front_fees
  def total_paid(self):
    return (years*payment_frequency)*payment_amount

  def current_interest(self, amount_paid):
    return ((principal - amount_paid)*interest_rate)/compounding
```

```python
class House:
  def __init__(self, value, annual_tax, current_owner, appreciation_rate):
    self.value = value
    self.annual_tax = annual_tax
    self.current_owner = current_owner
    self.appreciation_rate = appreciation_rate
  def forecast_value(self, years):
    appreciation_rate = self.appreciation_rate
    current_value = self.value
    for i in range(years):
      current_value = current_value + (appreciation_rate*current_value)
      print('after'+ str(i+1)+'years'+'value is'+str(current_value))
  def update_value(self, new_value):
    self.value = new_value
```

```python
home1 = House(800000,10000,'Bob',.2)
```

```python
home1.forecast_value(7)
```
