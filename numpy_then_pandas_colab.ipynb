We'll benefit from the following function:

```python
import re

def contains_non_numeric(string):
    pattern = r'[^0-9,.]'
    return re.search(pattern, string) is not None
```

```python
print(contains_non_numeric('fifty-five'))
print(contains_non_numeric('1,000,200'))
```

Using a Lambda function on multiple columns simultaneously:

```python
import pandas as pd

data = {'A': [1, 2, 3, 4, 5], 'B': [10, 20, 30, 40, 50]}
df = pd.DataFrame(data)

# Define the lambda function to operate on multiple columns
function = lambda row: row['A'] * 2 + row['B']

# Apply the lambda function to create the new column
df['C'] = df.apply(function, axis=1)

# Display the updated DataFrame
print(df)
```

```python
#using NumPy vectorize to do an opperation across an entire column
#using NumPy select to map multiple conditions to multiple different columns
import pandas as pd
import numpy as np

data = {'A': [1, 2, 3, 4, 5], 'B': [10, 20, 30, 40, 50]}
df = pd.DataFrame(data)

# Define the conditions
conditions = [
    df['A'] < 3,
    df['A'] >= 3
]

# Define the functions corresponding to each condition
def function1(x):
    return x * 2

def function2(x):
    return x + 5

# Vectorize the functions
vectorized_function1 = np.vectorize(function1)
vectorized_function2 = np.vectorize(function2)

# np.select is one of the most high-impact methods in numpy, because we map an array of conditions (booleans) to an array of values or functions.
# Apply the functions based on the conditions and create the new column
df['C'] = np.select(conditions, [vectorized_function1(df['A']), vectorized_function2(df['B'])], default=np.nan)

# Display the updated DataFrame
print(df)
```

```python
import pandas as pd
import numpy as np

data = {'A': [1, 2, 3, 4, 5], 'B': [10, 20, 30, 40, 50]}
df = pd.DataFrame(data)
df
# Define the conditions
conditions = [
    df['A'] < 3,
    df['A'] >= 3
]

# Define the functions corresponding to each condition
def function1(df):
    return df['A'] * 2 + df['B']

def function2(df):
    return df['A'] + df['B'] * 2

# Apply the functions based on the conditions and create the new column
df['C'] = np.select(conditions, [function1(df), function2(df)], default=np.nan)

# Display the updated DataFrame
print(df)
```

A non-trivial example of using lambda:

```python
footballers_goals = {'Eusebio': 120, 'Cruyff': 104, 'Pele': 150, 'Ronaldo': 132, 'Messi': 125, 'Bale':104}

sorted_footballers_by_goals = sorted(footballers_goals.items(), key= lambda x:(-x[1],x[0]),reverse = True)
sorted_footballers_by_goals
```

```python
footballers_goals.items()
```

```python
people_names = ['Joe Maxwell', 'Susan Grimes', 'Hiren Patel', 'Poonam Sharma', 'Will Edwards' , 'Zane Edwards', 'Adam Edwards']
sorted(people_names, key = lambda x: (x.split(' ')[1], x.split(' ')[0]))
```

```python
def custom_sort_key(item):
    return (-item[1], item[0])

footballers_goals = {'Eusebio': 120, 'Cruyff': 104, 'Pele': 150, 'Ronaldo': 132, 'Messi': 125, 'Bale': 104}

sorted_footballers_by_goals = sorted(footballers_goals.items(), key=custom_sort_key)
print(sorted_footballers_by_goals)
```
