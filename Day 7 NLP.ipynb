```python
def parse_elements(url,element):
    session = HTMLSession()
    r = session.get(url)
    elements = r.html.find(element)
    return elements
```

```python
def sentiment_detection(sentence):
    positive_words = ('happy', 'sunny', 'positive', 'triumphant', 'optimistic', 'wonderful', 'significant', 'achieve','capabilities','progress','prized')
    negative_words = ('sad', 'terrible', 'frightening', 'rainy', 'scary', 'shocked', 'critical','bottlenecked')

    positive = any(sentence.count(i) > 0 for i in positive_words)
    negative = any(sentence.count(i) > 0 for i in negative_words)

    if positive == negative == False:
        return "neutral"
    elif positive != negative:
        return "positive" if positive else "negative"
    else:
        return "mixed"
```

```python
def topic_detection(sentence):
    DataScience_words = ('data science', 'regression', 'predict', 'predictive', 'classify')
    AI_words = ('AI', 'machine learning', 'ml', 'ML', 'NLP', 'LLM','deep learning','neural network','generative', 'GPT-3')
    DataWords = ('analysis', 'algorithms','data','data structure','data-structure','big data')
    DataScience = any(sentence.count(i) > 0 for i in DataScience_words)
    AI = any(sentence.count(i) > 0 for i in AI_words)
    data = any(sentence.count(i) > 0 for i in DataWords)
    topics = []
    if DataScience == True:
        topics.append("Data Science")
    if AI == True:
        topics.append("AI")
    if data == True:
        topics.append("data")
    return topics
```

```python
def sentence_list(paragraph_string):
    tokenizer = nltk.data.load('tokenizers/punkt/english.pickle')
    list_of_sentences = tokenizer.tokenize(paragraph_string)
    return list_of_sentences
```

From this point onward we will work on repurposing our rules based functiosn that use AI to abstract away more of the process and using Pandas. We will see that switching from base Python to Pandas for those functions makes things easier.

```python
#requirements this will return the categories with the top 3 probabilities GIVEN that the probability is greater than 10% of the sum of the top 3
def topic_classifier(some_text):
  import pandas as pd
  inputs = tokenizer.encode_plus(some_text, padding="longest", truncation=True, return_tensors="pt")
  outputs = model(**inputs)
  #print(outputs)
  logits = outputs.logits
  probabilities = logits.softmax(dim=1)
  probabilities = list(probabilities[0])
  #with this next comprehension we just obtain the probability for each item, and not the tensor or other thing associated with it
  probabilities = [(probabilities[i]).item() for i in range(len(probabilities))]
  df = pd.DataFrame({'categories':['society and culture','Science', 'Health', 'education', 'Computers and Internet', 'sports', 'Finance', 'Entertainment and Music', 'Family and Relationships', 'Politics and Government'],'probabilities':probabilities})
  df.sort_values(by='probabilities', ascending=False, inplace=True)
  df = df.head(3)
  criterion = (df['probabilities'].sum())*.10
  return df[df['probabilities']>criterion]
```

```python
#repurpose with Pandas
#requirements: if max probability is greater than .6 return that
#requirements: if max probability less than .6 return all of those that have a value greater than .1
def classify_text_v2(some_text):
  import pandas as pd
  inputs = tokenizer.encode_plus(some_text, padding="longest", truncation=True, return_tensors="pt")
  outputs = model(**inputs)
  logits = outputs.logits
  probabilities = logits.softmax(dim=1)
  probabilities = list(probabilities[0])
  probabilities = [i.item() for i in probabilities]
  df = pd.DataFrame({'categories': ['society and culture','Science', 'Health', 'education', 'Computers and Internet', 'sports', 'Finance', 'Entertainment and Music', 'Family and Relationships', 'Politics and Government'], 'probabilities':probabilities})
  df.sort_values(by='probabilities', ascending=False, inplace=True)
  if df['probabilities'].max() > .6:
    return df[df['probabilities'] == df['probabilities'].max()]
  else:
    return df[df['probabilities']>0.1]
```
