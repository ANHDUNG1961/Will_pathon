```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create the DataFrame with three separate vectors for the horizontal axis
data1 = pd.DataFrame({
    'X1': [1, 2, 3, 4, 5, 6],
    'Y': [10, 15, 12, 17, 13, 16]
})

data2 = pd.DataFrame({
    'X2': [2, 4, 6, 8, 10, 12],
    'Y': [11, 17, 14, 12, 19, 10]
})


data3 = pd.DataFrame({
    'X3': [3, 6, 9, 12, 15, 18],
    'Y': [17, 13, 18, 12, 12, 18]
})




# Plotting the scatter plot
plt.scatter(data1['X1'], data1['Y'], color='red', label='Trend 1')
plt.scatter(data2['X2'], data2['Y'], color='green', label='Trend 2')
plt.scatter(data3['X3'], data3['Y'], color='blue', label='Trend 3')

# Customize the plot
plt.xlabel('Horizontal Axis')
plt.ylabel('Vertical Axis')
plt.title('Scatter Plot with Multiple Trends')
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Sample data (replace with your data)
x = country_data['Impartial.Public.Administration']
y = country_data['Economic.Freedom.Summary.Index']
z = country_data['Government.consumption']

# Fit a plane to the data
A = np.column_stack([x, y, np.ones_like(x)])
coefficients, residuals, _, _ = np.linalg.lstsq(A, z, rcond=None)

# Create a figure and a 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Scatter plot of the data points
ax.scatter(x, y, z)

# Create a meshgrid for the plane
x_plane, y_plane = np.meshgrid(np.linspace(min(x), max(x), 100), np.linspace(min(y), max(y), 100))
z_plane = coefficients[0] * x_plane + coefficients[1] * y_plane + coefficients[2]

# Plot the plane
ax.plot_surface(x_plane, y_plane, z_plane, alpha=0.5, cmap='viridis')

# Customize the plot
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')
ax.set_title('3D Scatter Plot with Plane of Best Fit')

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


# Generate random data for x and z
x = np.random.randn(500)
z = np.random.randn(500)

xbins = 20  # Number of bins in the x direction
zbins = 20  # Number of bins in the z direction

hist, xedges, zedges = np.histogram2d(x, z, bins=(xbins, zbins))

xmesh, zmesh = np.meshgrid(xedges, zedges)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the 3D histogram as a surface plot
ax.plot_surface(xmesh, zmesh, hist.T, cmap='viridis')

# Set labels for the axes
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Frequency')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate random data for x and z
x = np.random.randn(1000)
z = np.random.randn(1000)

# Define the number of bins
xbins = 20
zbins = 20

# Create a 2D histogram
hist, xedges, zedges = np.histogram2d(x, z, bins=(xbins, zbins))

# Create mesh grids for x and z
xmesh, zmesh = np.meshgrid(xedges, zedges)

# Create a figure and a 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Convert mesh grids to 1D arrays
xpos, zpos = xmesh.ravel(), zmesh.ravel()
bottom = np.zeros_like(xpos)
width = depth = 0.5

# Plot the 3D bar plot
ax.bar3d(xpos, zpos, bottom, width, depth, hist.ravel(), shade=True, cmap='viridis')

# Set labels for the axes
ax.set_xlabel('X Axis')
ax.set_ylabel('Z Axis')
ax.set_zlabel('Frequency')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate random data for x and z
x = np.random.randn(1000)
z = np.random.randn(1000)

# Define the number of bins
xbins = 20
zbins = 20

# Create a 3D histogram
hist, xedges, zedges = np.histogramdd((x, z), bins=(xbins, zbins))

# Create mesh grids for x and z
xmesh, zmesh = np.meshgrid(xedges, zedges)

# Create a figure and a 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

xpos, zpos = xmesh[:-1, :-1].ravel(), zmesh[:-1, :-1].ravel()
dx = dy = 0.25  # Adjust this for the bar width

# Plot the 3D bar plot
ax.bar3d(xpos, zpos, bottom=np.zeros_like(xpos), width=dx, depth=dy, height=hist.ravel(), shade=True, cmap='viridis')

# Set labels for the axes
ax.set_xlabel('X Axis')
ax.set_ylabel('Z Axis')
ax.set_zlabel('Frequency')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.stats import gaussian_kde

# Generate random data for x and z
x = np.random.randn(1000)
z = np.random.randn(1000)

# Create a 2D kernel density estimate
kde = gaussian_kde([x, z])

# Create a grid of x and z values for the density plot
x_grid, z_grid = np.mgrid[x.min():x.max():100j, z.min():z.max():100j]
positions = np.vstack([x_grid.ravel(), z_grid.ravel()])

# Evaluate the density at each grid point
density = kde(positions)

# Reshape the density values to match the shape of the grid
density = density.reshape(x_grid.shape)

# Create a figure and a 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the 3D density surface
ax.plot_surface(x_grid, z_grid, density, cmap='viridis')

# Set labels for the axes
ax.set_xlabel('X Axis')
ax.set_ylabel('Z Axis')
ax.set_zlabel('Density')

plt.show()
```
