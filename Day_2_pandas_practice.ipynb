We can do vectorized operations in base-python as well.

```python
test_list = ['yes', 'no', 'no', 'no', 'yes','yes','yes']
```

```python
list(map(lambda x: 1 if x == 'yes' else 0,test_list))
```

```python
def convert_to_dummy(x):
  if x == 'yes':
    x = 1
  else:
    x = 0
  return x
```

```python
list(map(convert_to_dummy,test_list))
```
