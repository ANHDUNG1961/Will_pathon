```python
#One solution to merge-the-tools
string = 'sdffgfasdwqw'
k = 3
sub_str = ''
for i in string:
  sub_str = sub_str + i
  if len(sub_str)%k == 0:
    new_str = ''
    for i in sub_str:
      if i not in new_str:
        new_str = new_str + i
    print(new_str)
    sub_str = ''
```

```python
def merge_the_tools(string, k):
    for i in range(0,len(string),k):
        subStr = string[i:i+k]
        print( ''.join(dict.fromkeys(subStr)) )
```
